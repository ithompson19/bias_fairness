# How to Run

The first code block contains the necessary imports to run tests and the selected data set (defaults to Adult data set). To run the analysis for any group, run the first code block and then run the code block for the corresponding analysis. To switch the data set from the Adult data set, change the `data_reader` variable in the first code block.

```python
import analyzer
from data_reader import Adult, Debug, SmallAdultEven, SmallAdultProp

# To switch data set, change this variable
data_reader = Adult
```
